### 3. Write a python program to implement gradient descent. 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
	m_curr = b_curr = 0
	iterations = 1000
	learning_rate = 0.01

	plt.scatter(x,y,color='blue', marker='o', linewidths=5)
	for i in range(iterations):
		y_pred = m_curr * x + b_curr
		plt.xlabel('x')
		plt.ylabel('y_predict')
		plt.plot(x, y_pred, color='green')

		cost = (sum([val for val in ((y-y_pred)**2)]))/len(x) #mean squared error. Formula: sum of (y-y_pred)^2 / n  or mathematicallly: 1/n * sum((y-y_pred)^2)
		
		# Partial Derivatives
		md = -(2/len(x)) * sum(x*(y-y_pred))
		bd = -(2/len(x)) * sum(y-y_pred)

		m_curr = m_curr - learning_rate * md
		b_curr = b_curr - learning_rate * bd

		print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
	
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

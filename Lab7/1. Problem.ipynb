```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def gaussian_distribution(x, mean, variance):
    exponent = -0.5 * ((x - mean) ** 2) / variance
    coefficient = 1 / np.sqrt(2 * np.pi * variance)
    return coefficient * np.exp(exponent)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
means = [0, 1, -2]
for mean in means:
    pdf = gaussian_distribution(x, mean, 1)  # Variance fixed at 1
    plt.plot(x, pdf, label=f"Mean={mean}")
    
```

```python
variances = [0.5, 1, 2]
for variance in variances:
    pdf = gaussian_distribution(x, 0, variance)  # Mean fixed at 0
    plt.plot(x, pdf, label=f"Variance={variance}")
    
```

```python
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Gaussian Distribution")
plt.legend()
plt.show()
```
